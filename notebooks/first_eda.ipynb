```python
import json
import pandas as pd
```

```python
# Caminho do arquivo JSON
file_path = "../data/raw/a1.json"

# Abrir e carregar o JSON
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)  # Carrega como dicionário
```

```python
# Converter para DataFrame (cada chave vira uma coluna)
df = pd.json_normalize(data).set_index(keys="mal_id")
```

```python
drop_cols = [
    'url', 'approved', 'images.jpg.image_url', 'status', 'airing', 'duration', 
       'images.jpg.small_image_url', 'images.jpg.large_image_url',
       'images.webp.image_url', 'images.webp.small_image_url',
       'images.webp.large_image_url', 'trailer.youtube_id', 'trailer.url',
       'trailer.embed_url', 'trailer.images.image_url',
       'trailer.images.small_image_url', 'trailer.images.medium_image_url',
       'trailer.images.large_image_url', 'trailer.images.maximum_image_url',
       'aired.from', 'aired.to', 'aired.prop.from.day',
       'aired.prop.from.month', 'aired.prop.from.year', 'aired.prop.to.day',
       'aired.prop.to.month', 'aired.prop.to.year',
       'broadcast.day', 'broadcast.time', 'broadcast.timezone', 'titles', 'title_english', 'title_japanese', 'members', 'favorites', 'explicit_genres', 'title_synonyms', 'background', 'scored_by'
]

df = df.drop(columns=drop_cols)
```

```python
df.info()
```

```python
df
```

```python
# Extrair apenas o campo 'name' da coluna e retornar como string separada por vírgulas
df['producers'] = df['producers'].apply(lambda x: ', '.join([producer['name'] for producer in x]) if isinstance(x, list) else '')
df['licensors'] = df['licensors'].apply(lambda x: ', '.join([licensor['name'] for licensor in x]) if isinstance(x, list) else '')
df['studios'] = df['studios'].apply(lambda x: ', '.join([studio['name'] for studio in x]) if isinstance(x, list) else '')
df['genres'] = df['genres'].apply(lambda x: ', '.join([genre['name'] for genre in x]) if isinstance(x, list) else '')
df['themes'] = df['themes'].apply(lambda x: ', '.join([theme['name'] for theme in x]) if isinstance(x, list) else '')
```

```python
df.to_csv("../data/processed/a1.csv", index=True, encoding="utf-8")
```
